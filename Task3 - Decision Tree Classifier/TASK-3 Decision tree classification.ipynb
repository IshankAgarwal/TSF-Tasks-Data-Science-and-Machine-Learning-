```python
# import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

### Read Data

```python
#read csv
data=pd.read_csv('iris.csv')
data.head()
```

```python
#filter if needed
data=data.drop(columns=['Id'])
data.head()
```

### Visualize the data

```python
#plotting
sns.pairplot(data,hue='Species',diag_kind='kde')
```

#### Extract Features and Labels for Training

```python
# Sepearting Features and target values
features =  data.iloc[:,0:4]
target =    data.iloc[:,[4]]

print(features.head())
print(target.head())
```

```python
#unique target classes
target_classes=data.Species.unique()
target_classes
```

### DecisionTree Classifier

```python
# implement Decision Tree
from sklearn.tree import DecisionTreeClassifier,tree
from sklearn import tree
model=DecisionTreeClassifier()
model.fit(features,target)
print(model )
```

### Visualize the Decision Tree

```python
#Plot Decision tree

#data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
plt.figure(figsize = (20,12))
tree.plot_tree(model,filled=True,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=target_classes)
plt.show()
```

### Predicting with a random sample 

```python
#predicting a random sample 
x=model.predict([[4.6, 3.1, 1.5, 0.2]])
print("Predicted output according to given sample is : ",x[0])
```
