```python
#data_target.values[:,0]

#extracted values from given dataset

val = data.iloc[:, [0, 1, 2, 3]].values
val
```

## Finding optimum number of clusters
Using Elbow method- The optimum clusters is where the elbow occurs

```python
#apply kmeans
from sklearn.cluster import KMeans
error=[]
# applying kmeans with different number of clusters
for i in range(1,12):
    kmeans_model=KMeans(n_clusters=i,random_state=0)
    kmeans_model.fit(val)
    error.append(kmeans_model.inertia_)
print(error)
plt.plot(range(1,12),error,c='r')
plt.grid()
plt.show()
```

According to above graph, elbow is at <b>3</b>, So optimum number of clusters = <b>3</b>.<br>
Therefore, Applying K-Means for n_clusters = <b>3</b>

```python
#According to Above graph and elbow method , best number of clustering =3 
#So, now fitting at clusters=3

kmeans_model_3=KMeans(n_clusters=3,random_state=0)
fit=kmeans_model_3.fit(val)

cent=(kmeans_model_3.cluster_centers_)     #centers
result=(fit.predict(val))
print(result)          #predicted labels
cent
```

### Visualizing clusters and their centers

```python
#visualizing data on different features and with clusterng centers

plt.scatter(val[:,0],val[:,1],c=result,cmap='rainbow')
plt.scatter(cent[:,0],cent[:,1],c='k',marker=',',s=100)
plt.grid()
plt.xlabel(data_feat.columns[0])
plt.ylabel(data_feat.columns[1])
plt.show()
plt.scatter(val[:,0],val[:,2],c=result,cmap='rainbow')
plt.scatter(cent[:,0],cent[:,2],c='k',marker=',',s=100)
plt.grid()
plt.xlabel(data_feat.columns[0])
plt.ylabel(data_feat.columns[2])
plt.show()
plt.scatter(val[:,1],val[:,2],c=result,cmap='rainbow')
plt.scatter(cent[:,1],cent[:,2],c='k',marker=',',s=100)
plt.grid()
plt.xlabel(data_feat.columns[1])
plt.ylabel(data_feat.columns[2])
plt.show()
plt.scatter(val[:,1],val[:,3],c=result,cmap='rainbow')
plt.scatter(cent[:,1],cent[:,3],c='k',marker=',',s=100)
plt.grid()
plt.xlabel(data_feat.columns[1])
plt.ylabel(data_feat.columns[3])
plt.show()
```
